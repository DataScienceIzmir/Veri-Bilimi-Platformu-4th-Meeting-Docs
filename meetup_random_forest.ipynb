```python
x = data[['peak-rpm','width','stroke','height','compression-ratio','num-of-cylinders','length',
      'wheel-base','horsepower','curb-weight','num-of-doors']]
y = data['price'].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)


from sklearn.ensemble import RandomForestRegressor

model2 = RandomForestRegressor(n_estimators=10, max_features=3,random_state=0)  
model2.fit(x_train, y_train) 
Y_prediction = model2.predict(x_test)

model2.score(x_train, y_train)
acc_random_forest = round(model2.score(x_train, y_train) * 100, 2)
print(round(acc_random_forest,2,), "%")

```

```python
from sklearn.metrics import r2_score
from scipy.stats import spearmanr, pearsonr

predicted_train = model2.predict(x_train)
predicted_test = model2.predict(x_test)

test_score = r2_score(y_test, predicted_test)

print "Test data R-2 score: {} ".format(test_score)
```

```python
# Extract single tree
estimator = model2.estimators_[5]

from sklearn.tree import export_graphviz
# Export as dot file
export_graphviz(estimator, out_file='tree.dot', 
                rounded = True, proportion = False, 
                filled = True)

# Convert to png using system command (requires Graphviz)
from subprocess import call
call(['dot', '-Tpng', 'tree.dot', '-o', 'tree.png', '-Gdpi=30'])

# Display in jupyter notebook
from IPython.display import Image
Image(filename = 'tree.png')
```
